```python
df['bmi'].fillna(28.893237, inplace=True)
```

```python
dumy = df["gender"]
df["gender"] = pd.get_dummies(dumy) # 0 for male , 1  for female
```

```python
df.head()
```

```python
df["smoking_status"] = df["smoking_status"].str.replace('Unknown',"never smoked").astype(str)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

ftr_list = ["work_type","smoking_status","Residence_type"]
encoder = LabelEncoder()
for ftr in ftr_list:
    df[ftr] = encoder.fit_transform(df[ftr])
    print(encoder.classes_)
    
df
```

```python
df.head()
```

```python
x = df.loc[:,["gender","age","heart_disease","work_type","Residence_type","avg_glucose_level","bmi","smoking_status"]].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=  train_test_split(x,y,test_size=0.4,random_state=5)
```

# Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
cl = DecisionTreeClassifier(criterion="entropy",random_state=0)
cl.fit(X_train,Y_train)
```

```python
cl.score(X_test,Y_test)
```

# Neighbors

```python
from sklearn import neighbors
knn=neighbors.KNeighborsClassifier()
knn.fit(X_train,Y_train)
```

```python
knn.score(X_test,Y_test)
```

# prediction with 94% score

```python
knn.predict(X_test)
```
